```python
def remove_values(df, col_name, sigma=3.0):
    mean = df[col_name].mean()
    std = df[col_name].std()
    up = mean + sigma*std
    down = mean - sigma*std
    df.drop(df[df[col_name] < down].index, inplace=True)
    df.drop(df[df[col_name] > up].index, inplace=True)
```

```python
for col in df.drop(columns=["diagnosis"]).columns:
    remove_values(df, col)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) 
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)

```

```python
X_train_norm
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score

#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=43)      
clr_rf = clf_rf.fit(X_train, y_train)

ac = accuracy_score(y_test, clf_rf.predict(X_test))
print('Accuracy is: ',ac)
```

```python
cm = confusion_matrix(y_test,clf_rf.predict(X_test))
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# find best scored 8 features
selector = SelectKBest(chi2, k=5)
z = selector.fit_transform(X_train, y_train)
```

```python
filter = selector.get_support()
features = X_train.columns
 
print("All features:")
print(features)
 
print("Selected best 5:")
print(features[filter])
print(z) 
 
```

Best 5 features to classify are: ''texture_mean', 'area_mean', 'concavity_mean', 'area_se', 'concavity_worst'.
Let's see what happens if we use only these best scored 5 feature.

```python
x_train_2 = selector.transform(X_train)
x_test_2 = selector.transform(X_test)

#random forest classifier with n_estimators=10 (default)
clf_rf_2 = RandomForestClassifier()      
clr_rf_2 = clf_rf_2.fit(x_train_2,y_train)
ac_2 = accuracy_score(y_test,clf_rf_2.predict(x_test_2))
print('Accuracy is: ',ac_2)
cm_2 = confusion_matrix(y_test,clf_rf_2.predict(x_test_2))
sns.heatmap(cm_2,annot=True,fmt="d")
```

```python
from sklearn.feature_selection import RFE

clf = RandomForestClassifier()
rfe = RFE(estimator=clf, n_features_to_select=5, step=1)
rfe.fit(X_train,y_train)
```

```python
print('Chosen best 5 feature by rfe:',X_train.columns[rfe.support_])
```

Chosen 5 best features by rfe is texture_mean, area_mean, concavity_mean, area_se, concavity_worst. They are exactly similar with previous (selectkBest) method. Therefore we do not need to calculate accuracy again.

Sklearn provides a similar RFECV class which implements Recursive Feature Elimination with cross-validation 
and automatically finds the optimal number of features to keep. Let's compare the results.

```python
from sklearn.feature_selection import RFECV

clf = RandomForestClassifier(random_state=42)
rfecv = RFECV(estimator=clf, step=1)
rfecv.fit(X_train,y_train)
```

```python
X_train.columns[rfecv.support_]
```

```python
X_train_new = rfecv.transform(X_train)
X_test_new = rfecv.transform(X_test)
```

```python
rf = RandomForestClassifier(random_state=42)
_ = rf.fit(X_train_new, y_train)

print("Trainign R-sqaured:", rf.score(X_train_new, y_train))
print("Testing R-squared:", rf.score(X_test_new, y_test))
```
