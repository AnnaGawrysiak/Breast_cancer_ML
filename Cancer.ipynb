```python
threshold = 0.85
corr_matrix = df.corr().abs()

df_triu = pd.DataFrame(np.triu(corr_matrix, k=1), columns=df.drop(columns=["diagnosis"]).columns)
columns = df_triu.columns
```

```python
columns = df_triu.columns
```

```python
plt.figure(figsize=(60, 60))
sns.heatmap(df_triu, annot=True)
```

```python
threshold = 0.85
high_correlated_features = [col for col in df_triu.columns if any(df_triu[col] > threshold)]
```

```python
high_correlated_features
```

```python
df = df.drop(columns = high_correlated_features)
```

```python
corr_matrix = df.corr()
```
